```python
import geopandas as gpd
import pandas as pd
```

```python
gdf = gpd.read_file('petitecouronne.geojson')
```

```python
gdf
```

```python
gdf.plot()
```

```python
gdf['geometry'] = gdf['geometry'].translate(-5.5,1.2)
```

```python
gdf['geometry'] = gdf['geometry'].scale(3,3)
```

```python
gdf[gdf['code'] == '75'].to_file('gdf75.geojson',driver='GeoJSON')
gdf[gdf['code'] == '92'].to_file('gdf92.geojson',driver='GeoJSON')
gdf[gdf['code'] == '93'].to_file('gdf93.geojson',driver='GeoJSON')
gdf[gdf['code'] == '94'].to_file('gdf94.geojson',driver='GeoJSON')
```

```python
gdf75 = gpd.read_file('gdf75.geojson')
gdf92 = gpd.read_file('gdf92.geojson')
gdf93 = gpd.read_file('gdf93.geojson')
gdf94 = gpd.read_file('gdf94.geojson')
```

```python
gdf92['geometry'] = gdf92['geometry'].translate(-0.21,-0.039)
gdf92.to_file('gdf92.geojson',driver='GeoJSON')
```

```python
gdf93['geometry'] = gdf93['geometry'].translate(0.2,0.1)
gdf93.to_file('gdf93.geojson',driver='GeoJSON')
```

```python
gdf94['geometry'] = gdf94['geometry'].translate(0.23,-0.17)
gdf94.to_file('gdf94.geojson',driver='GeoJSON')
```

```python
gdf = pd.concat([gdf75,gdf92,gdf93,gdf94])
```

```python
gdf
```

```python
gdf = gdf.rename(columns={'code':'dep','nom':'libgeo'})
```

```python
gdf['reg'] = '11'
```

```python
gdf.to_file('petite-couronne-scale-translate.geojson',driver='GeoJSON')
```
